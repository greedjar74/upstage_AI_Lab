### 3-3. ë³€ìˆ˜ ê°„ ê´€ê³„ í™•ì¸

#### ğŸ“ ì„¤ëª…: ìƒê´€ê³„ìˆ˜ì™€ ìƒê´€í–‰ë ¬

ìƒê´€ê³„ìˆ˜ëŠ” ë‘ ë³€ìˆ˜ê°€ í•¨ê»˜ ì›€ì§ì´ëŠ” ê²½í–¥ì„±, ì¦‰ ìƒê´€ê´€ê³„ì˜ ê°•ë„ë¥¼ -1ì—ì„œ 1ì‚¬ì´ì˜ ìˆ˜ì¹˜ë¡œ ë‚˜íƒ€ë‚¸
ì§€í‘œì…ë‹ˆë‹¤. ìƒê´€í–‰ë ¬(correlation matrix)ì€ ë°ì´í„°ì…‹ì— í¬í•¨ëœ ëª¨ë“  ë³€ìˆ˜ë“¤ì˜ ìŒì— ëŒ€í•´ì„œ
ìƒê´€ê³„ìˆ˜ë¥¼ ì¸¡ì •í•˜ê³  ì´ë¥¼ í–‰ë ¬ë¡œ ë‚˜íƒ€ë‚¸ ê²ƒì…ë‹ˆë‹¤.

ìƒê´€ê³„ìˆ˜ëŠ” ë‘ ë³€ìˆ˜ì˜ ìˆœì„œë¥¼ ë°”ê¾¸ì–´ë„ ë³€í•˜ì§€ ì•Šìœ¼ë¯€ë¡œ ìƒê´€í–‰ë ¬ì€ ëŒ€ì¹­í–‰ë ¬ë¡œ ë‚˜íƒ€ë‚˜ê³ , ìƒê´€í–‰ë ¬ì—ì„œ
ëŒ€ê°ì›ì†ŒëŠ” ìê¸° ìì‹ ê³¼ì˜ ìƒê´€ê³„ìˆ˜ë¡œ ë¬´ì¡°ê±´ 1.0ì´ ë˜ì–´ ì˜ë¯¸ê°€ ì—†ëŠ” ìˆ˜ì¹˜ì…ë‹ˆë‹¤. ë”°ë¼ì„œ ìƒê´€í–‰ë ¬ ë° ê·¸ì˜ ì‹œê°í™”ë¥¼ ë³¼ ë•ŒëŠ” í•˜ì‚¼ê°(lower-triangular)í–‰ë ¬ë§Œì„ í™•ì¸í•©ë‹ˆë‹¤.

ë‹¤ìŒ ì½”ë“œë¥¼ í†µí•´ ì§€ê¸ˆê¹Œì§€ ë¶„ì„í•œ ìœ íŠœë¸Œ í†µê³„ ë°ì´í„°ì˜ ìƒê´€í–‰ë ¬ì„ ì‹œê°í™”í•´ë´…ë‹ˆë‹¤.

- **NOTE**: ì›ë˜ëŠ” ìƒê´€ë¶„ì„ë„ ê²°ì¸¡ì¹˜ ì œê±° ë“±ì˜ ì „ì²˜ë¦¬ë¥¼ ê±°ì¹œ í›„ì— ìˆ˜í–‰í•´ì•¼í•˜ì§€ë§Œ ì—¬ê¸°ì—ì„œëŠ” ì‹¤ì œ
ë¶„ì„ë³´ë‹¤ëŠ” ë‚´ìš© ì„¤ëª…ì´ ëª©ì ì´ë¯€ë¡œ ê·¸ëƒ¥ ì§„í–‰í•©ë‹ˆë‹¤.

````python
# ë°ì´í„°ì˜ ìƒê´€ê³„ìˆ˜ í–‰ë ¬ì„ ìƒì„±
corr = data.corr(numeric_only=True)  # ë°ì´í„°ì…‹ìœ¼ë¡œë¶€í„° ê° ë³€ìˆ˜ë³„ ìƒê´€ê³„ìˆ˜ë¥¼ ê³„ì‚°í•˜ì—¬ í–‰ë ¬í˜•íƒœë¡œ ë‚˜íƒ€ë‚´ëŠ” í•¨ìˆ˜. ì—¬ê¸°ì„œëŠ” í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ë¥¼ ì´ìš©
# numeric_onlyëŠ” ìˆ«ìí˜• ë³€ìˆ˜ë§Œ ìƒê´€ê³„ìˆ˜ë¥¼ ê³„ì‚°í•˜ë„ë¡ í•˜ëŠ” ì˜µì…˜
# figureì—ì„œ ìƒëµë  ë¶€ë¶„ì„ ì§€ì •í•˜ëŠ” mask í–‰ë ¬ì„ ìƒì„± - ì‹¤ì œë¡œëŠ” mask ì—†ì´ ë‹¤ ìƒì„±í•´ë„ ê´œì°®ìŠµë‹ˆë‹¤.
mask = np.ones_like(corr, dtype=bool)
mask = np.triu(mask)

# ì‹œê°í™”ë  ê·¸ë¦¼ì˜ í¬ê¸°ë¥¼ ì§€ì •
# íˆíŠ¸ë§µ í˜•íƒœë¡œ ìƒê´€í–‰ë ¬ ì‹œê°í™”í•˜ê¸°
plt.figure(figsize=(13,10))
sns.heatmap(data=corr, annot=True, fmt='.2f', mask=mask, linewidths=.5, cmap='RdYlBu_r')
plt.title('Correlation Matrix')
plt.show()
````

ìƒê´€í–‰ë ¬ ê·¸ë¦¼ì„ ë³´ë©´, ì§„í•œ ë¹¨ê°„ìƒ‰ìœ¼ë¡œ ë‚˜íƒ€ë‚˜ëŠ” ë³€ìˆ˜ ìŒì´ ì¡´ì¬í•©ë‹ˆë‹¤. ìƒê´€ê³„ìˆ˜ê°€ 1ì— ê°€ê¹Œìš´ ìˆ«ìë¡œ ë†’ê²Œ ë‚˜íƒ€ë‚ ìˆ˜ë¡ ì–‘ì˜ ìƒê´€ê´€ê³„ê°€ ì¡´ì¬í•¨ì„ ì˜ë¯¸í•©ë‹ˆë‹¤. ì˜ˆë¥¼ ë“¤ì–´, êµ¬ë…ììˆ˜(subscribers)ì™€ ì¡°íšŒìˆ˜(video view)ëŠ” 0.75ì˜ ìƒê´€ê³„ìˆ˜ ê°’ì„ ê°€ì§€ë©°, ì–‘ì˜ ìƒê´€ê´€ê³„ê°€ ì¡´ì¬í•¨ì„ ì•Œ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì¦‰, êµ¬ë…ììˆ˜ê°€
ë§ì„ìˆ˜ë¡ ì˜ìƒì˜ ì¡°íšŒìˆ˜ê°€ ëŒ€ì²´ì ìœ¼ë¡œ ë†’ë‹¤ëŠ” ì •ë³´ë¥¼ íŒŒì•…í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ë°˜ë©´ì—, ì§„í•œ íŒŒë€ìƒ‰ìœ¼ë¡œ ë‚˜íƒ€ë‚˜ëŠ” ë³€ìˆ˜ ìŒë„ ì¡´ì¬í•©ë‹ˆë‹¤. ì´ ë³€ìˆ˜ë“¤ì€ ìƒê´€ê³„ìˆ˜ê°€ -1ì— ê°€ê¹Œìš´ ìˆ«ìë¡œ ë‚®ê²Œ ë‚˜íƒ€ë‚˜, ìŒì˜ ìƒê´€ê´€ê³„ê°€ ì¡´ì¬í•¨ì„ ì˜ë¯¸í•©ë‹ˆë‹¤. ì˜ˆë¥¼ ë“¤ì–´, êµ¬ë…ììˆ˜(subscribers)ì™€ ìˆœìœ„(rank)ëŠ” -0.65ì˜ ìƒê´€ê³„ìˆ˜ ê°’ì„ ê°€ì§€ë©°, ì´ëŠ” ë‘ ë³€ìˆ˜ ê°„ ìŒì˜ ìƒê´€ê´€ê³„ê°€ ì¡´ì¬í•¨ì„ ì•Œ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì¦‰, êµ¬ë…ììˆ˜ê°€ ë§ì„ìˆ˜ë¡ ëŒ€ì²´ì ìœ¼ë¡œ ìˆœìœ„ê°€ ë†’ì•„ rank ìˆ«ìê°€ ë‚®ìŠµë‹ˆë‹¤.

ìƒê´€ê³„ìˆ˜ê°€ ì•„ì˜ˆ 1.0ì´ë‚˜ -1.0ì¸ ë³€ìˆ˜ë“¤ì€, ì‹¤ì œë¡œ ë§¤ìš° ê°•í•œ ìƒê´€ê´€ê³„ë¥¼ ê°€ì§€ëŠ” ì˜ë¯¸ìˆëŠ” ë³€ìˆ˜ì¼ ê°€ëŠ¥ì„±ì´
ì—†ì§€ëŠ” ì•Šì§€ë§Œ, ê·¸ë³´ë‹¤ëŠ” **ë…ë¦½ì ìœ¼ë¡œ ë‹¤ë£° ìˆ˜ ì—†ëŠ” ë³€ìˆ˜**ì¼ ê°€ëŠ¥ì„±ì´ í›¨ì”¬ ë†’ìŠµë‹ˆë‹¤.

ì˜ˆë¥¼ ë“¤ì–´ "ìˆ˜ìµ"ì´ë¼ëŠ” í•œ ë³€ìˆ˜ê°€ í•œ columnì—ì„œëŠ” ë‹¬ëŸ¬ ë‹¨ìœ„ë¡œ, ë˜ë‹¤ë¥¸ columnì—ì„œëŠ” ì›í™” ë‹¨ìœ„ë¡œ ì¤‘ë³µë˜ì–´ í‘œê¸°ë˜ì–´ìˆë‹¤ë©´ ë‘ ë³€ìˆ˜ ì‚¬ì´ì˜ ìƒê´€ê³„ìˆ˜ëŠ” 1ë¡œ ë‚˜íƒ€ë‚˜ê² ì§€ë§Œ ì‹¤ì œë¡œ í•œ ë³€ìˆ˜ë¥¼ í†µí•´ ë‹¤ë¥¸ í•˜ë‚˜ë¥¼ ì˜ˆì¸¡í•˜ëŠ” ë“±ì˜ ë¶„ì„ì€ ì „í˜€ ì˜ë¯¸ì—†ëŠ” ì¼ì´ ë  ê²ƒì…ë‹ˆë‹¤.
ì´ë ‡ê²Œ í•œ ë³€ìˆ˜ê°€ ë‹¤ë¥¸ ë³€ìˆ˜ì™€ ì‚¬ì‹¤ìƒ ê°™ì€ ê°’ìœ¼ë¡œ í‘œí˜„ë˜ì–´ë²„ë¦¬ëŠ” ê²½ìš° ë‘ ë³€ìˆ˜ëŠ” **ê³µì„ ì„±(Colinearity)**ì´ ìˆë‹¤ê³  í•˜ë©°, ë¹„ìŠ·í•˜ê²Œ í•œ ë³€ìˆ˜ê°€ í‰ê· ê°’ ë“± ë‹¤ë¥¸ ì—¬ëŸ¬ ë³€ìˆ˜ì˜ ì¡°í•©ìœ¼ë¡œ ë‚˜íƒ€ë‚˜ëŠ” ê²½ìš°ëŠ” **ë‹¤ì¤‘ ê³µì„ ì„±(Multicolinearity)**ì´ ìˆë‹¤ê³  ë¶€ë¦…ë‹ˆë‹¤.

ìƒê´€í–‰ë ¬ì—ì„œ ìƒê´€ê³„ìˆ˜ê°€ 1.0 ë˜ëŠ” -1.0ìœ¼ë¡œ ë‚˜íƒ€ë‚˜ëŠ” ë³€ìˆ˜ë“¤ì€ ì´ë ‡ê²Œ ê³µì„ ì„±ì´ë‚˜ ë‹¤ì¤‘ê³µì„ ì„±ì„ ê°€ì§„ ë³€ìˆ˜ì¼ ê°€ëŠ¥ì„±ì´ í¬ê¸° ë•Œë¬¸ì— ì£¼ì˜ê¹Šê²Œ ì‚´í´ë³´ì•„ì•¼ í•©ë‹ˆë‹¤.

### 4-2. ë°ì´í„° ë¶„í¬ ë³€í™˜

#### ğŸ“ ì„¤ëª…: ë¡œê·¸ ë³€í™˜

ì•ì—ì„œ ë³€ìˆ˜ 'lowest_monthly_earnings' ì˜ ë¶„í¬ê°€ ì˜¤ë¥¸ìª½ìœ¼ë¡œ ë„ˆë¬´ ê¼¬ë¦¬ê°€ ê¸¸ì–´ ì™¼ìª½ìœ¼ë¡œ ì¹˜ìš°ì³ì§„ ê²ƒì„ í™•ì¸í–ˆìŠµë‹ˆë‹¤. ì´ë¥¼ ë³´ì •í•˜ì—¬ ë¶„í¬ë¥¼ ë³´ê¸° ìœ„í•´ log scaleë¡œ ë³€í™˜ì„ í–ˆëŠ”ë°, ì´í›„ ë³€ìˆ˜ì˜ ì¹˜ìš°ì¹¨ì´ ëœí•´ì¡ŒìŠµë‹ˆë‹¤.

ì´ì²˜ëŸ¼ ìš°ë¦¬ê°€ ì›í•˜ëŠ” ë¶„í¬ì˜ í˜•íƒœë¡œ ê´€ì¸¡ëœ ë°ì´í„°ì˜ __ë¶„í¬ë¥¼ ë³€í™˜__í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì´ëŸ¬í•œ ì¢…ë¥˜ì˜ ì²˜ë¦¬ëŠ”
ë°ì´í„° ë¶„ì„ ê³¼ì •ì— ë„ì›€ì´ ë ìˆ˜ ìˆì§€ë§Œ, ê·¸ë ‡ì§€ ì•Šê³  ì•…ì˜í–¥ì„ ë¼ì¹˜ëŠ” ê²½ìš°ë„ ìƒê¸¸ ìˆ˜ ìˆì–´, ë³€ìˆ˜ë¥¼ í•´ì„í• 
ë•ŒëŠ” ë³€í™˜ì„ ì ìš©í–ˆë‹¤ëŠ” ì‚¬ì‹¤ì„ ì¶©ë¶„íˆ ê³ ë ¤í•˜ì—¬ì•¼í•˜ë©°, í•„ìš”í•œ ê²½ìš° ë³€í™˜ ì´ì „ì˜ ë°ì´í„°ë¡œ ë˜ëŒë ¤ ì‚¬ìš©í•©ë‹ˆë‹¤.


````python
# ë°ì´í„° ë¶„í¬ ë³€í™˜ì„ ìœ„í•œ scikit-learn íŒ¨í‚¤ì§€ ë‚´ í•¨ìˆ˜ ë¶ˆëŸ¬ì˜¤ê¸°
from sklearn.preprocessing import scale
````

ë‹¤ìŒê³¼ ê°™ì´ ê¼¬ë¦¬ê°€ ê¸¸ì–´ ì¹˜ìš°ì¹œ ë¶„í¬(long-tail, skewd)ë¥¼ ì •ê·œë¶„í¬ì— ê°€ê¹ê²Œ í•˜ê¸° ìœ„í•´ log scaleë¡œ ë³€í™˜í•©ë‹ˆë‹¤.

````python
# ë¡œê·¸ë³€í™˜ì„ ì§„í–‰í•  ë³€ìˆ˜ë¥¼ ì„ íƒí•©ë‹ˆë‹¤. ì—¬ê¸°ì—ì„œ ë‹¤ë¥¸ í”¼ì³ë¥¼ ì§€ì •í•˜ë©´ ë‹¤ë¥¸ í”¼ì³ì—ë„ ë¡œê·¸ë³€í™˜ì„ ì§„í–‰í•´ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
target_feature = 'lowest_monthly_earnings'
# ë°ì´í„° ë¡œê·¸ ë³€í™˜
# ë¡œê·¸ ìŠ¤ì¼€ì¼ë¡œ ë³€í™˜í•œ ê°’ìœ¼ë¡œ ìƒˆ ë³€ìˆ˜ë¥¼ ë§Œë“­ë‹ˆë‹¤. ìƒˆ ë³€ìˆ˜ëª…ì€ ê¸°ì¡´ ë³€ìˆ˜ëª…ì— prefixë¥¼ ë¶™ì…ë‹ˆë‹¤.
data[f'log_{target_feature}'] = scale(np.log(data[target_feature]+1))  # logì— 0ê°’ì´ ë“¤ì–´ê°€ëŠ” ê²ƒì„ í”¼í•˜ê¸° ìœ„í•´ 1ì„ ë”í•´ì¤ë‹ˆë‹¤.
# ë¡œê·¸ë³€í™˜í•œ ë³€ìˆ˜ì˜ ìš”ì•½í†µê³„ëŸ‰ ë³´ê¸°
data[f'log_{target_feature}'].describe()
````

### 4-3. ë°ì´í„° ë‹¨ìœ„ ë³€í™˜

#### ğŸ“ ì„¤ëª…: ì •ê·œí™”ì™€ í‘œì¤€í™”
ë³€ìˆ˜ `lowest_monthly_earnings`ëŠ” 0ì—ì„œ 850,900ê¹Œì§€ì˜ ìˆ˜ì¹˜ë¥¼ ê°€ì§€ê³  ìˆì§€ë§Œ ë³€ìˆ˜ `rank`ëŠ” 1ì—ì„œ 995ê¹Œì§€ì˜ ìˆ˜ì¹˜ë¥¼ ê°€ì§‘ë‹ˆë‹¤. ë³€ìˆ˜ `lowest_monthly_earnings`ì—ê²Œ 100 ì •ë„ì˜ ìˆ˜ì¹˜ ì°¨ì´ëŠ”
ì ì„ ìˆ˜ ìˆì§€ë§Œ, ë³€ìˆ˜ `rank`ì—ì„œ 100 ì •ë„ëŠ” ë§¤ìš° ë‹¤ë¥¸ ìˆ˜ì¹˜ì…ë‹ˆë‹¤. ë°ì´í„°ì˜ ë¶„ì„ê³¼ì •ì—ì„œ ì´ëŸ° ìŠ¤ì¼€ì¼ì˜
ì°¨ì´ëŠ” ëª¨ë¸ì˜ í•™ìŠµì´ ëŠë¦¬ê²Œ ë§Œë“¤ê±°ë‚˜, ì˜ˆìƒì¹˜ëª»í•œ ì—ëŸ¬ë¥¼ ì¼ìœ¼í‚¤ëŠ” ë“±ì˜ ë¶€ì •ì ì¸ ì˜í–¥ì„ ë¼ì¹  ìˆ˜ ìˆìœ¼ë¯€ë¡œ
ë¶„ì„ ì „ì— ê° ë³€ìˆ˜ë¥¼ ë¹„ìŠ·í•œ ìŠ¤ì¼€ì¼ë¡œ ë³€í™˜í•´ì£¼ëŠ” ê²ƒì´ ì¢‹ìŠµë‹ˆë‹¤.
ë‹¤ë§Œ, ìœ„ì˜ ë¡œê·¸ë³€í™˜ ì˜ˆì‹œì™€ ë‹¬ë¦¬ ë¶„í¬ì˜ ëª¨ì–‘ì„ ë³€í™˜ì‹œí‚¤ì§€ëŠ” ì•ŠìŠµë‹ˆë‹¤.

ì—¬ê¸°ì—ì„œëŠ” ë°ì´í„° ë‚´ ê° ë³€ìˆ˜ì˜ ìŠ¤ì¼€ì¼ì„ ê· ë“±í•˜ê²Œ ë§Œë“¤ê¸° ìœ„í•´, í‘œì¤€í™”(Standardization)ì™€ ì •ê·œí™”(Normalization)ë¥¼ í†µí•œ ë°ì´í„° ë‹¨ìœ„ ë³€í™˜ì„ ì‚¬ìš©í•´ë³´ê² ìŠµë‹ˆë‹¤.

- **NOTE**: Normalization, Standardizationì€ ëª¨ë‘ í•œêµ­ì–´ë¡œ "ì •ê·œí™”"ë¼ê³  ë²ˆì—­ë  ìˆ˜ ìˆê³ ,
ë§¥ë½ì— ë”°ë¼ ì—¬ê¸°ì—ì„œ ì •ì˜ëœê²ƒê³¼ ë‹¤ë¥´ê²Œ ì‚¬ìš©ë˜ëŠ” ë“± ìš©ì–´ì˜ í˜¼ë™ì´ ìˆëŠ” ìš©ì–´ì…ë‹ˆë‹¤.

**í‘œì¤€í™”(Standardization)**ëŠ” ë°ì´í„°ì˜ ë¶„í¬ë¥¼ í‰ê· ì´ 0ì´ê³  ë¶„ì‚°ì´ 1ì´ ë˜ë„ë¡ ë³€í™˜í•˜ëŠ” ê²ƒìœ¼ë¡œ, ë‹¤ìŒ ìˆ˜ì‹ìœ¼ë¡œ í‘œí˜„í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

$x_\textrm{scaled}=\frac{x-\textrm{Mean}(\mathbf{x})}{\textrm{Stdv}(\mathbf{x})}$.

ì—¬ê¸°ì„œ $x$ëŠ” ë³€í™˜í•˜ê³ ì í•˜ëŠ” ì›ë˜ ë°ì´í„° í¬ì¸íŠ¸ ê°’, $\mathbf{x}$ëŠ” í•œ ë³€ìˆ˜ì˜ ë°ì´í„° ê°’ë“¤ (ë²¡í„°), ê·¸ë¦¬ê³  $\textrm{Mean}(\mathbf{x})ê³¼ \textrm{Stdv}(\mathbf{x})$ëŠ” ë²¡í„° $\mathbf{x}$ì˜ í‰ê· ê³¼ í‘œì¤€í¸ì°¨ë¥¼ ì˜ë¯¸í•©ë‹ˆë‹¤.

**ì •ê·œí™”(Normalization í˜¹ì€ minmax scaling)**ëŠ” ë°ì´í„°ë¥¼ ìµœì†Ÿê°’ì´ 0ì´ê³  ìµœëŒ“ê°’ì´ 1ë¡œ, 0ì—ì„œ
1ì‚¬ì´ì˜ ê°’ì„ ê°€ì§€ë„ë¡ ë³€í™˜í•˜ëŠ” ë°©ë²•ì´ë©° ë‹¤ìŒ ì‹ìœ¼ë¡œ í‘œí˜„ë©ë‹ˆë‹¤.

$x_\textrm{scaled}=\frac{x-\textrm{Min}(\mathbf{x})}{\textrm{Max}(\mathbf{x})-\textrm{Min}(\mathbf{x})}$.

ì—¬ê¸°ì„œ $\textrm{Min}(\mathbf{x})$ê³¼ $\textrm{Max}(\mathbf{x})$ëŠ” ë²¡í„° $\mathbf{x}$ì˜ ìµœì†Ÿê°’ê³¼ ìµœëŒ“ê°’ì„ ì˜ë¯¸í•©ë‹ˆë‹¤.

````python
# ë°ì´í„° ë¶„í¬ ë³€í™˜ì„ ìœ„í•œ scikit-learn íŒ¨í‚¤ì§€ ë‚´ í•¨ìˆ˜ ë¶ˆëŸ¬ì˜¤ê¸°
from sklearn.preprocessing import StandardScaler, MinMaxScaler
````

ë¨¼ì €, í‘œì¤€í™”(Standardization)ì„ ì´ìš©í•˜ì—¬ ë°ì´í„°ì˜ ë‹¨ìœ„ë¥¼ ë³€í™˜í•´ë³´ê² ìŠµë‹ˆë‹¤.

````python
# ìŠ¤ì¼€ì¼ë§ì„ ì§„í–‰í•  ë³€ìˆ˜ë¥¼ ì„ íƒí•©ë‹ˆë‹¤. ì—¬ê¸°ì—ì„œ ë‹¤ë¥¸ í”¼ì³ë¥¼ ì§€ì •í•˜ë©´ ë‹¤ë¥¸ í”¼ì³ì—ë„ í‘œì¤€í™”ë¥¼ ì§„í–‰í•´ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
target_feature = 'lowest_monthly_earnings'
# scikit-learníŒ¨í‚¤ì§€ë¥¼ í™œìš©í•œ standardization
standard_scaler = StandardScaler()
data[f'standardized_{target_feature}'] = standard_scaler.fit_transform(data[[target_feature]])  # í‘œì¤€í™”í•œ ë°ì´í„°ë¥¼ ìƒˆë¡œìš´ ë³€ìˆ˜ì— ì €ì¥í•˜ê¸°
````

````python
# ë¹„êµë¥¼ ìœ„í•´ ë³€í™˜ ì „í›„ì˜ í”¼ì³ë¥¼ ì„ íƒ
feature_original = data[target_feature]
feature_standardized = data[f'standardized_{target_feature}']

# ë³€í™˜ ì „í›„ì˜ í‰ê· , í‘œì¤€í¸ì°¨ ë¹„êµ
print(f"í‰ê· (mean)ë¹„êµ: {feature_original.mean():.7} -> {feature_standardized.mean():.7}")
print(f"í‘œì¤€í¸ì°¨(standard deviation)ë¹„êµ: {feature_original.std():.7} -> {feature_standardized.std():.7}")
````

ë³€í™˜ í›„ ë°ì´í„° ìš”ì•½ í†µê³„ëŸ‰ì„ í™•ì¸í•´ë³´ë©´, í‰ê· (mean)ê³¼ í‘œì¤€í¸ì°¨(std)ê°€ ê°ê° 0, 1ì˜ ê°’ìœ¼ë¡œ ë³€í™˜ëœ ê²ƒì„ í™•ì¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
í•˜ì§€ë§Œ í‘œì¤€í™”ì™€ ì •ê·œí™”ëŠ” ë¶„í¬ì˜ ëª¨ì–‘ ìì²´ë¥¼ ë³€í™”ì‹œí‚¤ëŠ”ê²ƒì€ ì•„ë‹ˆë¯€ë¡œ ê·¸ë˜í”„ì˜ ëª¨ì–‘ì€ ì™„ì „íˆ ë™ì¼í•˜ê²Œ ë‚˜íƒ€ë‚©ë‹ˆë‹¤.

````python
# ë°ì´í„° ë‹¨ìœ„ ë³€í™˜ ì „í›„ ë¹„êµ
fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(15,6))  # ë°ì´í„° ë‹¨ìœ„ ë³€í™˜ ì „í›„ë¥¼ ë¹„êµí•˜ê¸° ìœ„í•´ 2ê°œì˜ plot ì§€ì •
# íˆìŠ¤í† ê·¸ë¨ìœ¼ë¡œ ë³€í™˜ ì „ ë°ì´í„°ì˜ ë¶„í¬ ë³´ê¸°
sns.histplot(data=data, x=target_feature, kde=True, ax=ax[0])
ax[0].set_title('Before standardization')
# íˆìŠ¤í† ê·¸ë¨ìœ¼ë¡œ ë³€í™˜ í›„ ë°ì´í„°ì˜ ë¶„í¬ ë³´ê¸°
sns.histplot(data=data, x=f'standardized_{target_feature}', kde=True, ax=ax[1])
ax[1].set_title('After standardization')
plt.show()
````

ë‹¤ìŒìœ¼ë¡œ, ì •ê·œí™”(Normalization)ëŠ” ë‹¤ìŒê³¼ ê°™ì´ ìˆ˜í–‰í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
# ìŠ¤ì¼€ì¼ë§ì„ ì§„í–‰í•  ë³€ìˆ˜ë¥¼ ì„ íƒí•©ë‹ˆë‹¤. ì—¬ê¸°ì—ì„œ ë‹¤ë¥¸ í”¼ì³ë¥¼ ì§€ì •í•˜ë©´ ë‹¤ë¥¸ í”¼ì³ì—ë„ í‘œì¤€í™”ë¥¼ ì§„í–‰í•´ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
target_feature = 'lowest_monthly_earnings'
# scikit-learníŒ¨í‚¤ì§€ë¥¼ í™œìš©í•œ normalization
normalized_scaler = MinMaxScaler()
data[f'normalized_{target_feature}'] = normalized_scaler.fit_transform(data[[target_feature]])  # í‘œì¤€í™”í•œ ë°ì´í„°ë¥¼ ìƒˆë¡œìš´ ë³€ìˆ˜ì— ì €ì¥í•˜ê¸°
````

````python
# ë¹„êµë¥¼ ìœ„í•´ ë³€í™˜ ì „í›„ì˜ í”¼ì³ë¥¼ ì„ íƒ
feature_original = data[target_feature]
feature_normalized = data[f'normalized_{target_feature}']

# ë³€í™˜ ì „í›„ì˜ ìµœëŒ€, ìµœì†Œê°’ ë¹„êµ
print(f"ìµœì†Œê°’(min) ë¹„êµ: {feature_original.min():.7} -> {feature_normalized.min():.7}")
print(f"ìµœëŒ€ê°’(max) ë¹„êµ: {feature_original.max():.7} -> {feature_normalized.max():.7}")
````
