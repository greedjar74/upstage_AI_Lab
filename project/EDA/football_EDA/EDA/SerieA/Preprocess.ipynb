## 포지션 전처리

```python
SA_Player_df['position'].unique()
```

```python
# 전체 데이터에서 position 데이터만 추출

position_df = SA_Player_df[['position']]
position_df
```

```python
# 리스트 형태로 변환

tmp_li = position_df['position'].tolist()
tmp_li
```

```python
# 포지션이 여러 개인 경우 하나의 포지션만 남긴다.

for i in range(len(tmp_li)):
    tmp = tmp_li[i]
    if ',' in tmp:
        tmp_li[i] = tmp.split(',')[0]
```

```python
tmp = pd.DataFrame(tmp_li)
tmp.columns = ['position']
tmp
```

```python
# 전체 데이터에 새로운 포지션 값 적용

SA_Player_df['position'] = tmp['position']
SA_Player_df['position'] # 변경이 되었는지 확인
```

```python
# unique값 확인
# 공백이 있는 경우 같은 포지션도 다르게 판단 -> 공백을 지워준다.

tmp_li = SA_Player_df['position'].tolist()
tmp_li
```

```python
# 공백 삭제

for i in range(len(tmp_li)):
    tmp_li[i] = tmp_li[i].replace(' ', '')
tmp_li
```

```python
tmp = pd.DataFrame(tmp_li, columns=['position'])
tmp
```

```python
# 원래 데이터에 적용

SA_Player_df['position'] = tmp
```

```python
# 각 포지션이 잘 적용되었는지 확인

SA_Player_df['position'].unique()
```

## 전처리 추가 필요 -> FW, AM, M, DM, D, GK

```python
# 포지션 압축
# M, FW, AM, D, DMC, M, GK, Forward

tmp_df = SA_Player_df['position']
tmp_li = tmp_df.tolist()
tmp_li
```

```python
# 포지션 압축

for i in range(len(tmp_li)):
    pos_tmp = tmp_li[i]
    if '(' in pos_tmp:
        pos_tmp_li = pos_tmp.split('(')
        tmp_li[i] = pos_tmp_li[0]
    
    if tmp_li[i] == 'Forward':
        tmp_li[i] = 'FW'
        
    elif tmp_li[i] == 'DMC':
        tmp_li[i] = 'DM'
        
    elif tmp_li[i] == 'Midfielder':
        tmp_li[i] = 'M'
```

```python
# 원본 데이터에 적용

SA_Player_df['position'] = pd.DataFrame(tmp_li, columns=['position'])
```
