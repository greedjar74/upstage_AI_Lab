# GK 포지션 데이터
## GK포지션에서 승리에 가장 큰 영향을 주는 변수 추출
- 포지션의 특수성을 생각해서 필요없는 변수는 미리 제거

```python
SA_GK_df = pd.merge(SA_GK_player_df, SA_Team_df[['Team', 'Pts']], on='Team', how='inner')
```

```python
SA_GK_df.columns
```

```python
# 수치형 데이터만 뽑아온다.

SA_GK_numeric_df = SA_GK_df.select_dtypes(['int64', 'float64'])
```

```python
SA_GK_corr_matrix = SA_GK_numeric_df.corr()
SA_GK_corr_matrix
```

```python
# heatmap으로 상관관계 시각화

plt.figure(figsize=(10, 10))
sns.heatmap(SA_GK_corr_matrix)
plt.show()
```

```python
# 결과가 nan인 변수 제거

SA_GK_df.drop(['Crosses', 'ThrB', 'Goals', 'Red', 'SpG', 'Off', 'Disp', 'Offsides', 'Drb', 'Blocks'], axis=1, inplace=True)
```

```python
# 수치형 데이터만 뽑아온다.

SA_GK_numeric_df = SA_GK_df.select_dtypes(['int64', 'float64'])
```

```python
SA_GK_corr_matrix = SA_GK_numeric_df.corr()
SA_GK_corr_matrix
```

```python
# heatmap으로 상관관계 시각화

plt.figure(figsize=(10, 10))
sns.heatmap(SA_GK_corr_matrix)
plt.show()
```

```python
# Pts부분만 뽑아서 상관관계 확인

SA_GK_corr_matrix_Pts = SA_GK_corr_matrix[['Pts']]
SA_GK_corr_matrix_Pts
```

```python
# 내림차순 정렬 후 heatmap으로 시각화

SA_GK_corr_matrix_Pts.sort_values('Pts', ascending=False, inplace=True)
plt.figure(figsize=(3, 10))
sns.heatmap(SA_GK_corr_matrix_Pts, annot=True, fmt='.3f')
plt.show()
```

```python
# 상위 10개 변수 추출

SA_GK_corr_matrix_Pts.iloc[1:11].index.to_list()
# ['PS%', 'age', 'Fouls', 'Rating', 'Dribble', 'OwnG', 'Mins', 'Inter', 'Tackles', 'UnsTch']
```

```python
index_li = SA_GK_corr_matrix_Pts.iloc[:11].index.tolist()
corr_li = SA_GK_corr_matrix_Pts['Pts'].tolist()
for i in range(1, 11):
    print(f'{index_li[i]}: {corr_li[i]:.3f}')
    
# PS%: 0.740 age: 0.163 Fouls: 0.059 Rating: 0.019 Dribble: 0.007 OwnG: 0.004 Mins: -0.021 Inter: -0.028 Tackles: -0.082 UnsTch: -0.128
```
