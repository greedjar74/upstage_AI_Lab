## Laliga 팀 분석 데이터

# GK 포지션 데이터
## GK포지션에서 승리에 가장 큰 영향을 주는 변수 추출
- 포지션의 특수성을 생각해서 필요없는 변수는 미리 제거

```python
LL_GK_df = pd.merge(LL_GK_player_df, LL_Team_df[['team_name', 'Pts']], on='team_name', how='inner')
```

```python
# GK 데이터 cols 확인
# 필요없는 변수는 제거해준다.

LL_GK_df.columns
```

```python
# 수치형 데이터만 뽑아온다.

LL_GK_numeric_df = LL_GK_df.select_dtypes(['int64', 'float64'])
```

```python
LL_GK_corr_matrix = LL_GK_numeric_df.corr()
LL_GK_corr_matrix
```

```python
# heatmap으로 상관관계 시각화

plt.figure(figsize=(15, 15))
sns.heatmap(LL_GK_corr_matrix)
plt.show()
```

```python
# 결과가 nan인 변수 제거

LL_GK_df.drop(['Assists', 'Crosses', 'ThrB', 'Goals','Red', 'SpG', 'Off', 'Fouls', 'Offsides', 'Blocks', 'OwnG'], axis=1, inplace=True)
```

```python
# 수치형 데이터만 뽑아온다.

LL_GK_numeric_df = LL_GK_df.select_dtypes(['int64', 'float64'])
```

```python
LL_GK_corr_matrix = LL_GK_numeric_df.corr()
LL_GK_corr_matrix
```

```python
# heatmap으로 상관관계 시각화

plt.figure(figsize=(10, 10))
sns.heatmap(LL_GK_corr_matrix)
plt.show()
```

```python
# Pts부분만 뽑아서 상관관계 확인

LL_GK_corr_matrix_Pts = LL_GK_corr_matrix[['Pts']]
LL_GK_corr_matrix_Pts
```

```python
# 내림차순 정렬 후 heatmap으로 시각화

LL_GK_corr_matrix_Pts.sort_values('Pts', ascending=False, inplace=True)
plt.figure(figsize=(3, 10))
sns.heatmap(LL_GK_corr_matrix_Pts, annot=True, fmt='.3f')
plt.show()
```

```python
# 상위 10개 변수 추출

LL_GK_corr_matrix_Pts.iloc[1:11].index.to_list()
#['PS%', 'AvgP', 'KeyP', 'Rating', 'Dribble', 'age', 'Inter', 'Apps', 'Mins', 'Tackles']
```

```python
index_li = LL_GK_corr_matrix_Pts.iloc[:11].index.tolist()
corr_li = LL_GK_corr_matrix_Pts['Pts'].tolist()
for i in range(1, 11):
    print(f'{index_li[i]}: {corr_li[i]:.3f}')
    
# PS%: 0.503 AvgP: 0.344 KeyP: 0.279 Rating: 0.219 Dribble: 0.105 age: 0.096 Inter: 0.087 Apps: 0.018 Mins: 0.009 Tackles: -0.014
```
