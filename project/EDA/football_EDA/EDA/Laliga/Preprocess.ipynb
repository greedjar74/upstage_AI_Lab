## 포지션 전처리

```python
LL_Player_df['position'].unique()
```

```python
# 전체 데이터에서 position 데이터만 추출

position_df = LL_Player_df[['position']]
position_df
```

```python
# 리스트 형태로 변환

tmp_li = position_df['position'].tolist()
tmp_li
```

```python
# 포지션이 여러 개인 경우 하나의 포지션만 남긴다.

for i in range(len(tmp_li)):
    tmp = tmp_li[i]
    if ',' in tmp:
        tmp_li[i] = tmp.split(',')[0]
```

```python
tmp = pd.DataFrame(tmp_li)
tmp.columns = ['position']
tmp
```

```python
# 전체 데이터에 새로운 포지션 값 적용

LL_Player_df['position'] = tmp['position']
LL_Player_df['position'] # 변경이 되었는지 확인
```

```python
# unique값 확인
# 공백이 있는 경우 같은 포지션도 다르게 판단 -> 공백을 지워준다.

tmp_li = LL_Player_df['position'].tolist()
tmp_li
```

```python
# 공백 삭제

for i in range(len(tmp_li)):
    tmp_li[i] = tmp_li[i].replace(' ', '')
tmp_li
```

```python
tmp = pd.DataFrame(tmp_li, columns=['position'])
tmp
```

```python
# 원래 데이터에 적용

LL_Player_df['position'] = tmp
```

```python
# 각 포지션이 잘 적용되었는지 확인

LL_Player_df['position'].unique()
```

## Apps 전처리

```python
# Apps -> 출전횟수
# 교체 출전의 경우 괄호에 있다.
# 그냥 합쳐준다.
apps_df = LL_Player_df['Apps']
apps_df.head()
```

```python
tmp_li = apps_df.tolist()
tmp_li
```

```python
# ()안에 있는 숫자를 그냥 더해줌
# Apps값을 int형으로 변환

for i in range(343):
    if '(' in tmp_li[i]:
        tmp_li[i] = tmp_li[i].split('(')
        tmp_li[i] = [tmp_li[i][0], tmp_li[i][1].replace(')', '')]
        tmp_li[i] = int(tmp_li[i][0]) + int(tmp_li[i][1])
    else:
        tmp_li[i] = int(tmp_li[i])
    
tmp_df = pd.DataFrame(tmp_li, columns=['Apps'])
tmp_df
```

```python
LL_Player_df['Apps'] = tmp_df
LL_Player_df[['Apps']]
```

```python
# 추후 병합을 위해 team_name을 맞춰준다. -> team_table_added값을 기준으로

player_team_df = LL_Player_df['team_name']
player_team_list = player_team_df.to_list()


for i in range(len(player_team_list)): 
    player_team = player_team_list[i]
    
    if player_team == 'Atletico':
        player_team_list[i] = 'Atletico Madrid'
    elif player_team == 'Athletic Bilbao':
        player_team_list[i] = 'Athletic Club'
        

LL_Player_df['team_name'] = pd.DataFrame(player_team_list, columns=['team_name'])
LL_Player_df['team_name'].unique()
```

# 포지션 구분
## 전처리 추가 필요 -> FW, AM, M, DM, D, GK

```python
position_li = LL_Player_df['position'].unique().tolist()
position_li
```

```python
# 각 포지션별로 데이터 추출

Position_df_li = []

for position in position_li:
    tmp = LL_Player_df[LL_Player_df['position'] == position]
    Position_df_li.append(tmp)
```

```python
print(position_li[0])
Position_df_li[0]
```

```python
# 포지션 압축
# M, FW, AM, D, DMC, M, GK, Forward

tmp_df = LL_Player_df['position']
tmp_li = tmp_df.tolist()
tmp_li
```

```python
# 포지션 압축

for i in range(len(tmp_li)):
    pos_tmp = tmp_li[i]
    if '(' in pos_tmp:
        pos_tmp_li = pos_tmp.split('(')
        tmp_li[i] = pos_tmp_li[0]
    
    if tmp_li[i] == 'Forward':
        tmp_li[i] = 'FW'
        
    elif tmp_li[i] == 'DMC':
        tmp_li[i] = 'DM'
        
    elif tmp_li[i] == 'Midfielder':
        tmp_li[i] = 'M'
```

```python
# 원본 데이터에 적용

LL_Player_df['position'] = pd.DataFrame(tmp_li, columns=['position'])
```

```python
# 적용되었는지 확인

LL_Player_df['position'].unique()
```
