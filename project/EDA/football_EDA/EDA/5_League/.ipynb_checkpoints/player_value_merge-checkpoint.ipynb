```python
import pandas as pd
import os
```

```python
bl_player = pd.read_csv('./Budesliga_data/Bundesliga_Player_22_23.csv')
bl_player.drop('Unnamed: 0', axis=1, inplace=True)
bl_player
```

```python
player_list = bl_player['player_name'].unique().tolist()
```

```python
bl_path_list = os.listdir('./Budesliga_data/Player_Value/')
print(len(bl_path_list))
merged_li = []

for path in bl_path_list:
    tmp = pd.read_csv('./Budesliga_data/Player_Value/'+path)
    tmp.columns = ['player_name', 'Position', 'Date of Birth', 'Age', 'Nationality', 'Market Value']
    merged_li.append(pd.merge(bl_player, tmp, on='player_name', how='inner'))
```

```python
len(merged_li)
```

```python
bl_merged = pd.concat([merged_li[0], merged_li[5], merged_li[10], merged_li[14],
                       merged_li[1], merged_li[6], merged_li[11], merged_li[15],
                       merged_li[2], merged_li[7], merged_li[12], merged_li[16],
                       merged_li[3], merged_li[8], merged_li[13], merged_li[17],
                       merged_li[4], merged_li[9]], ignore_index=True)
```

```python
bl_merged
```
