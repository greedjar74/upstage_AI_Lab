# Apps 전처리

```python
# Apps -> 출전횟수
# 교체 출전의 경우 괄호에 있다.
# 그냥 합쳐준다.
apps_df = BL_player['Apps']

tmp_li = apps_df.tolist()

# ()안에 있는 숫자를 그냥 더해줌
# Apps값을 int형으로 변환

for i in range(BL_player.shape[0]):
    if '(' in tmp_li[i]:
        tmp_li[i] = tmp_li[i].split('(')
        tmp_li[i] = [tmp_li[i][0], tmp_li[i][1].replace(')', '')]
        tmp_li[i] = int(tmp_li[i][0]) + int(tmp_li[i][1])
    else:
        tmp_li[i] = int(tmp_li[i])
    
tmp_df = pd.DataFrame(tmp_li, columns=['Apps'])

BL_player['Apps'] = tmp_df
BL_player[['Apps']]
```

```python
# 추후 병합을 위해 Team이름을 맞춰준다.
player_team_list = BL_player['Team'].to_list()


for i in range(BL_player.shape[0]): 
    player_team = player_team_list[i]
    
    if player_team == 'SC Freiburg':
        player_team_list[i] = 'Freiburg'
    elif player_team == 'VfL Wolfsburg':
        player_team_list[i] = 'Wolfsburg'
    elif player_team == 'FSV Mainz':
        player_team_list[i] = 'Mainz 05'
    elif player_team == 'Borussia monchegladbach':
        player_team_list[i] = 'Borussia M.Gladbach'
    elif player_team == 'Koln':
        player_team_list[i] = 'FC Koln'
    elif player_team == 'TSG Hoffenheim':
        player_team_list[i] = 'Hoffenheim'
    elif player_team == 'Stuttgart':
        player_team_list[i] = 'VfB Stuttgart'
    elif player_team == 'Schalke':
        player_team_list[i] = 'Schalke 04'
    elif player_team == 'Union Berlin':
        player_team_list[i] = 'Hertha Berlin'
        

BL_player['Team'] = pd.DataFrame(player_team_list, columns=['Team'])
BL_player['Team'].unique()
```

```python
BL_team.columns = ['Team', 'P', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts', 'Goals',
       'Shots pg', 'Yellow', 'Red', 'Poss%', 'Pass%', 'A_Won', 'Rating',
       'Shoted pg', 'Tackles pg', 'Intercept pg', 'Fouls pg', 'Offsides pg',
       'Shots OT pg', 'Dribbles pg', 'Fouled pg']
```
