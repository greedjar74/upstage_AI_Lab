# GK 포지션 데이터
## GK포지션에서 승리에 가장 큰 영향을 주는 변수 추출
- 포지션의 특수성을 생각해서 필요없는 변수는 미리 제거

```python
BL_GK_df = pd.merge(BL_GK_player_df, BL_Team_df[['Team', 'Pts']], on='Team', how='inner')
```

```python
BL_GK_df.columns
```

```python
# 수치형 데이터만 뽑아온다.

BL_GK_numeric_df = BL_GK_df.select_dtypes(['int64', 'float64'])
```

```python
BL_GK_corr_matrix = BL_GK_numeric_df.corr()
BL_GK_corr_matrix
```

```python
# heatmap으로 상관관계 시각화

plt.figure(figsize=(10, 10))
sns.heatmap(BL_GK_corr_matrix)
plt.title('GK values Corr')
plt.show()
```

```python
# 결과가 nan인 변수 제거

BL_GK_df.drop(['Assists', 'Crosses', 'ThrB', 'Goals', 'SpG', 'Off', 'Offsides', 'Blocks'], axis=1, inplace=True)
```

```python
# 수치형 데이터만 뽑아온다.

BL_GK_numeric_df = BL_GK_df.select_dtypes(['int64', 'float64'])
```

```python
BL_GK_corr_matrix = BL_GK_numeric_df.corr()
BL_GK_corr_matrix
```

```python
# heatmap으로 상관관계 시각화

plt.figure(figsize=(10, 10))
sns.heatmap(BL_GK_corr_matrix)
plt.title('GK values Corr')
plt.show()
```

```python
# Pts부분만 뽑아서 상관관계 확인

BL_GK_corr_matrix_Pts = BL_GK_corr_matrix[['Pts']]
BL_GK_corr_matrix_Pts
```

```python
# 내림차순 정렬 후 heatmap으로 시각화

BL_GK_corr_matrix_Pts.sort_values('Pts', ascending=False, inplace=True)
plt.figure(figsize=(3, 10))
sns.heatmap(BL_GK_corr_matrix_Pts, annot=True, fmt='.3f')
plt.show()
```

```python
# 상위 5개 변수 추출

# 결과분석
# PS%: 패스 성공률 -> 음.. 조금 중요도는 낮을 것 같지만 의미있는 지표
# Dribble: 드리블..? -> 의미없는 지표
# Fouls: 음.. 약간 의미있음
# Disp: ?
# Raiting: 약간 의미있음

# 전체적으로 상관관계가 낮아서 정확한 지표로 사용하기에는 문제가 있음

BL_GK_corr_matrix_Pts.iloc[1:].index.to_list()
```

```python
index_li = BL_GK_corr_matrix_Pts.iloc[:11].index.tolist()
corr_li = BL_GK_corr_matrix_Pts['Pts'].tolist()
for i in range(1, 11):
    print(f'{index_li[i]}: {corr_li[i]:.3f}')
    
# PS%: 0.553 Dribble: 0.473 Fouls: 0.432 Disp: 0.432 Rating: 0.367 team_number: 0.279 OwnG: 0.157 age: 0.063 Red: 0.053 AerialsWon: 0.044
```
