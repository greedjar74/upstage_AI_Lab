# 데이터 전처리
- .dropna(subset='x'): subset에 설정한 x변수에 결측치가 있는 경우 해당 행을 삭제

### 남극 데이터 전처리
- 기온을 구할 수 없는 경우 행을 삭제 -> 최저, 최고 기온이 없는 경우
- 결측치 기온은 최고, 최저의 평균으로 대체한다.

```python
# 남극 데이터에서 기온, 최고 기온, 최저 기온이 모두 없는 경우를 찾는다.
# 기온에 대한 데이터가 모두 없으면 기온에 대한 데이터를 구할 수가 없다.
# 기온이 없으면서 최저 혹은 최고가 없는 경우 제거 -> 모두 없는 경우도 자연스럽게 제거됨

new_antatica.isna().sum()
```

```python
# 기온이 없으면서 최저 또는 최고 기온이 없는 인덱스를 구한다.

tmp = new_antatica[new_antatica['기온'].isna()] # 기온이 np.nan인 경우를 찾는다.

a = tmp[tmp['최저 기온'].isna()].index.tolist() # 최저 기온도 없는 경우를 찾는다.
b = tmp[tmp['최고 기온'].isna()].index.tolist() # 최고 기온도 없는 경우를 찾는다.

antatica_delete_index = list(set(a + b)) # 삭제할 인덱스를 구한다.
antatica_delete_index
```

```python
print(new_antatica.shape)
new_antatica[new_antatica['기온'].isna()]
```

```python
# 구할 수 없는 값인 경우 행 삭제

new_antatica = new_antatica.drop(antatica_delete_index, axis=0)
```

```python
# 기온이 결측치지만 값을 구할 수 있는 경우만 남았는지 확인한다.
new_antatica[new_antatica['기온'].isna()]
```

```python
# 행이 잘 삭제되었는지 확인

new_antatica.shape
```

```python
mask = new_antatica['기온'].isna() # 조건
tmp = (new_antatica[mask]['최고 기온'] + new_antatica[mask]['최저 기온']) / 2 # 결측치를 대체할 값을 구한다. -> 최고, 최저의 평균
change_idx = tmp.index.tolist() # 변경되야 되는 데이터의 인덱스

# 기온 결측치 처리
for idx in change_idx:
    new_antatica.loc[idx, '기온'] = tmp.loc[idx] # .loc[idx]['기온']을 하면 에러 발생

# 결측치가 사라졌는지 확인
new_antatica['기온'].isna().sum()
```

```python
# 기온 변화 그래프 시각화

plt.figure(figsize=(15, 10))
plt.plot(new_antatica['기온'], label='temp')
plt.grid(True)
plt.title('Antatica Temp')
plt.legend()
plt.show()
```

# 예측 모델 생성

### linear regression
- 다중선형회귀 모델 생성

```python
import statsmodels.api as sm
```

```python
# 학습, 테스트 데이터 분할 없이 전체 데이터를 학습에 사용

# x, y 데이터 생성
X_data = data.drop(['서울 기온'], axis=1)
y_data = data[['서울 기온']]

X_data = sm.add_constant(X_data, has_constant='add') # 상수항 추가

# 모델 생성 및 적합
linear_regression_model = sm.OLS(y_data, X_data)
fitted_linear_regression_model = linear_regression_model.fit()

print(fitted_linear_regression_model.summary()) # 모델 정보 출력 -> R-squared 값이 낮다... -> 예측 성능이 낮다...

# 변수들의 P값은 전체적으로 낮다. -> 변수들 사이의 상관계수가 낮아서 변수를 굳이 삭제할 필요는 없어 보인다.
```

```python
# 학습, 테스트 데이터를 분할해서 사용

from sklearn.model_selection import train_test_split

# x, y 변수 생성
X = data.drop(['서울 기온'], axis=1) # target인 '서울 기온'을 제외한 데이터를 모두 가져온다.
y = data[['서울 기온']] # '서울 기온'만 가져온다.

X = sm.add_constant(X, has_constant='add') # X 데이터에 상수항 추가

# 학습, 테스트 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=42)

print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
# 모델 생성 및 적합
linear_regression_model = sm.OLS(y_train, X_train)
fitted_linear_regression_model = linear_regression_model.fit()

print(fitted_linear_regression_model.summary()) # 모델 정보 출력 -> R-squared 값이 낮다... -> 예측 성능이 낮다...

# 변수들의 P값은 전체적으로 낮다. -> 변수들 사이의 상관계수가 낮아서 변수를 굳이 삭제할 필요는 없어 보인다.
```

```python
# 예측과 실제값 그래프 시각화

plt.figure(figsize=(15, 5))
plt.plot(np.array(fitted_linear_regression_model.predict(X_test)), label = 'pred')
plt.plot(np.array(y_test), label='true')
plt.legend()
plt.grid(True)
plt.title('Seoul Temperature Prediction by Linear Regression Model')
plt.show()
```
