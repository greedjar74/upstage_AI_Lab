```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
np.random.seed(42)
data_size=300
features = 10

x = np.random.randn(data_size, features)
y = (np.random.randn(data_size) > 0.5).astype(int)

columns = [f'feature{i}' for i in range(features)]
x_tmp = pd.DataFrame(x, columns=columns)
y_tmp = pd.DataFrame(y, columns=['label'])

df = pd.concat([x_tmp, y_tmp], axis=1)
df
```

```python
X = df.drop(['label'], axis=1)
y = df[['label']]

train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.8, test_size=0.2)

print(train_X.shape, test_X.shape)
print(train_y.shape, test_y.shape)
```

```python
scaler = StandardScaler()
train_X_scaled = scaler.fit_transform(train_X)
```

```python
train_X_scaled
```

```python
model = LogisticRegression()
model
```

```python
fitted_model = model.fit(train_X_scaled, train_y)
```

```python
test_X_scaled = scaler.fit_transform(test_X)
```

```python
pred_y = fitted_model.predict(test_X_scaled)
```

```python
pred_y_df = pd.DataFrame(pred_y, columns=['label'])
```

```python
test_y
```

```python
accuracy = accuracy_score(test_y, pred_y)
accuracy
```

```python
confusion_matrix(test_y, pred_y)
```

```python
print(classification_report(test_y, pred_y))
```
